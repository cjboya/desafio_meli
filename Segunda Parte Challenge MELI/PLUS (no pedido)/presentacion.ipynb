```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(r'C:\Users\Distancia\Desktop\martes1508\resolver_mercadolibre\Segunda Parte Challenge MELI\resultados.csv')
df


```

```python
df.info()
```

```python
# Resumen estadístico de las columnas numéricas
print(df.describe())
```

```python
# Gráfico de barras para mostrar la distribución de condiciones
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Condicion')
plt.title('Distribución de Condiciones')
plt.xlabel('Condición')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python


# Gráfico de barras para mostrar la distribución de ciudades
# Hay que mejorarlo, ya que algunos vendedores no ponen el barrio correctamente
# Y toma ciudad, de BsAs
# Pero se puede ya ver algo de información interesante.

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Ciudad', order=df['Ciudad'].value_counts().index[:10])
plt.title('Top 10 Ciudades')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()

```

```python
# Gráfico de histograma para mostrar la distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Precio', bins=20, kde=True)
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```
